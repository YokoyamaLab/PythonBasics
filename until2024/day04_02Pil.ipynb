### ğŸ’®ãƒãƒªã‚´ãƒ³

```python
# ç”»åƒã®å¤§ãã•ã¨èƒŒæ™¯è‰²ã‚’æ±ºã‚ã‚‹(è‰²ã¯èµ¤,ç·‘,é’ã®å¼·ã•ã‚’0ï½255ã§è¡¨ã™)
image = Image.new("RGBA", (400, 400), (0, 0, 255))

# å›³å½¢ã‚’æãã‚­ãƒ£ãƒ³ãƒã‚¹ã‚’æº–å‚™ã™ã‚‹
canvas = ImageDraw.Draw(image)

# ã‚­ãƒ£ãƒ³ãƒã‚¹ã«ç·šåˆ†ã‚’æç”»ã™ã‚‹
canvas.polygon(
    [(50, 100), (300, 350),(50, 350),(280, 20)],  # å·¦ä¸Šã®xyåº§æ¨™ã€å³ä¸‹ã®XYåº§æ¨™
    fill=(0,255,0),
    outline=(0,40,0)
)

# colabãƒšãƒ¼ã‚¸ä¸Šã«è¡¨ç¤ºã™ã‚‹
image
```

### ğŸ’®å¤šè§’å½¢

ã€[ãƒªãƒ•ã‚¡ãƒ¬ãƒ³ã‚¹](https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.regular_polygon)ã€‘

```python
# ç”»åƒã®å¤§ãã•ã¨èƒŒæ™¯è‰²ã‚’æ±ºã‚ã‚‹(è‰²ã¯èµ¤,ç·‘,é’ã®å¼·ã•ã‚’0ï½255ã§è¡¨ã™)
image = Image.new("RGBA", (400, 400), (0, 0, 255))

# å›³å½¢ã‚’æãã‚­ãƒ£ãƒ³ãƒã‚¹ã‚’æº–å‚™ã™ã‚‹
canvas = ImageDraw.Draw(image)

steps = 10
# foræ–‡ã§è¨­å®šã‚’å¤‰ãˆãªãŒã‚‰æ²¢å±±æã„ã¦ã¿ã‚‹
for i in range(steps):
  # ã‚­ãƒ£ãƒ³ãƒã‚¹ã«å¤šè§’å½¢ã‚’æç”»ã™ã‚‹
  canvas.regular_polygon(
      (200,200,200-(i*15)),
      n_sides = steps + 3 - i,
      rotation = i * 360 / steps,
      fill=(0,25 * i,0)
  )

# colabãƒšãƒ¼ã‚¸ä¸Šã«è¡¨ç¤ºã™ã‚‹
image
```

### ğŸ’®æ¥•å††ï¼å††

ã€[ãƒªãƒ•ã‚¡ãƒ¬ãƒ³ã‚¹](https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.ellipse)ã€‘

```python
# ç”»åƒã®å¤§ãã•ã¨èƒŒæ™¯è‰²ã‚’æ±ºã‚ã‚‹(è‰²ã¯èµ¤,ç·‘,é’ã®å¼·ã•ã‚’0ï½255ã§è¡¨ã™)
image = Image.new("RGBA", (400, 400), (0, 0, 255))

# å›³å½¢ã‚’æãã‚­ãƒ£ãƒ³ãƒã‚¹ã‚’æº–å‚™ã™ã‚‹
canvas = ImageDraw.Draw(image)

# ã‚­ãƒ£ãƒ³ãƒã‚¹ã«é•·æ–¹å½¢ã‚’æç”»ã™ã‚‹
canvas.rectangle(
    [(50, 100), (300, 350)],  # å·¦ä¸Šã®xyåº§æ¨™ã€å³ä¸‹ã®XYåº§æ¨™
    outline=(255, 255, 255),  # ç·šã®è‰²
    width=1                   # ç·šã®å¤ªã•
)

# ã‚­ãƒ£ãƒ³ãƒã‚¹ã«æ¥•å††ã‚’æç”»ã™ã‚‹ï¼ˆä¸Šè¨˜é•·æ–¹å½¢ã«å†…æ¥ã™ã‚‹å††ãŒæã‘ã‚‹ï¼‰
canvas.ellipse(
    [(50, 100), (300, 350)],  # å·¦ä¸Šã®xyåº§æ¨™ã€å³ä¸‹ã®XYåº§æ¨™
    outline=(0, 255, 0),      # ç·šã®è‰²
    fill=(255,0,0),           # å¡—ã‚Šã®è‰²
    width=3                   # ç·šã®å¤ªã•
)


# colabãƒšãƒ¼ã‚¸ä¸Šã«è¡¨ç¤ºã™ã‚‹
image
```

### ğŸ’®å††å¼§

å††å¼§ã¯ç«¯ç‚¹ã®å‡¦ç†æ–¹æ³•ã§ä¸‰ç¨®é¡ã‚ã‚Šã¾ã™ã€‚

ã€[ãƒªãƒ•ã‚¡ãƒ¬ãƒ³ã‚¹: arc](https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.arc)ã€‘

ã€[ãƒªãƒ•ã‚¡ãƒ¬ãƒ³ã‚¹: chord](https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.chord)ã€‘

ã€[ãƒªãƒ•ã‚¡ãƒ¬ãƒ³ã‚¹: pieslice](https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.pieslice)ã€‘

```python
# ç”»åƒã®å¤§ãã•ã¨èƒŒæ™¯è‰²ã‚’æ±ºã‚ã‚‹(è‰²ã¯èµ¤,ç·‘,é’ã®å¼·ã•ã‚’0ï½255ã§è¡¨ã™)
image = Image.new("RGBA", (400, 400), (0, 0, 255))

# å›³å½¢ã‚’æãã‚­ãƒ£ãƒ³ãƒã‚¹ã‚’æº–å‚™ã™ã‚‹
canvas = ImageDraw.Draw(image)
coordinates = [(50, 100), (300, 350)]

# ã‚­ãƒ£ãƒ³ãƒã‚¹ã«é•·æ–¹å½¢ã‚’æç”»ã™ã‚‹
canvas.rectangle(
    coordinates,  # å·¦ä¸Šã®xyåº§æ¨™ã€å³ä¸‹ã®XYåº§æ¨™
    outline=(255, 255, 255),  # ç·šã®è‰²
    width=1                   # ç·šã®å¤ªã•
)

# èµ¤ãŒarc
canvas.arc(
    coordinates,  # å·¦ä¸Šã®xyåº§æ¨™ã€å³ä¸‹ã®XYåº§æ¨™
    start=5,
    end=115,
    fill=(255,0,0),
    width=3
)

# ç·‘ãŒchord
canvas.chord(
    coordinates,
    start=125,
    end=235,
    fill=(0,255,0),
    outline=(0,120,0),
    width=3
)

# é»„è‰²ãŒpieslice
canvas.pieslice(
    coordinates,  # å·¦ä¸Šã®xyåº§æ¨™ã€å³ä¸‹ã®XYåº§æ¨™
    start=245,
    end=355,
    fill=(255,255,0),
    outline=(120,120,0),
    width=3
)
# colabãƒšãƒ¼ã‚¸ä¸Šã«è¡¨ç¤ºã™ã‚‹
image
```
