```python
from google.colab import auth
auth.authenticate_user()
import gspread
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

filename = "Day2 èª²é¡Œ02 å‰æº–å‚™ï¼ˆå›ç­”ï¼‰"
file = gc.open(filename)
sheet = file.get_worksheet(0)
table = sheet.get_all_values()
# print(table)

```

ã“ã‚Œã§ã€çš†ã•ã‚“ã®ãƒãƒ¼ãƒˆã«ã‚¢ãƒ³ã‚±ãƒ¼ãƒˆçµæœãŒèª­ã¿ã“ã¾ã‚Œã¾ã—ãŸã€‚

æ¬¡ã®åˆ†æã§ã¯ä½•ã‚’ã™ã‚Œã°ã‚ˆã„ã§ã—ã‚‡ã†ã‹ã€‚è²¡å¸ƒã®å€¤æ®µã¨å°éŠ­ã®é¡ã®ãƒ‡ãƒ¼ã‚¿ãŒã‚ã‚Šã¾ã™ã®ã§ã€ã“ã‚Œã‚’æ•£å¸ƒå›³ã«ã—ã¦è¡¨ç¤ºã—ã¦ã¿ã¾ã™ã€‚

```python
# ä¸€è¡Œç›®ã¯åˆ—åæ–‡å­—åˆ—ãªã®ã§æŠœã„ã¦ãŠã
table.pop(0)
# é‡‘é¡ã‚‚æ–‡å­—åˆ—ã¨ã—ã¦å…¥ã£ã¦ã„ã‚‹ã®ã§intã«å¤‰æ›ã—ã¦ãŠã
table = list(map(lambda x: (x[0],x[1],int(x[2]),int(x[3]),x[4],x[5]),table))
# tableã‚’Pandasã®ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã«å¤‰æ›
dataframe = pd.DataFrame.from_records(table)
# print(dataframe)

dataframe.plot.scatter(x=2, y=3, xlabel="Amount of coins", ylabel="Price of wallets")
```

å‡ºã¦æ¥ã¾ã—ãŸã€‚

ã“ã‚Œã ã‘ã§ã¯ã€ç›®ã§è¦‹ã¦ã€Œãµãƒ¼ã‚“ã€ã¨ã„ã†äº‹ã—ã‹ã§ãã¾ã›ã‚“ã€‚ãƒ‡ãƒ¼ã‚¿ã‚µã‚¤ã‚¨ãƒ³ã‚¹çš„ã«ã¯ã€ã“ã“ã«ä½•ã‹ä»®èª¬ã‚’æ¤œè¨¼ã™ã‚‹ãŸã‚ã®æƒ…å ±ã‚’åŠ ãˆã¦ã¿ã¾ã™ã€‚

ã“ã“ã§èª¿ã¹ãŸã„ã®ã¯ã€è²¡å¸ƒã®å€¤æ®µã¨å°éŠ­ã®é¡ãŒæ¯”ä¾‹é–¢ä¿‚ã«ã‚ã‚‹ã‹ã©ã†ã‹ã§ã™ã®ã§ã€å›å¸°ç›´ç·šã‚’å¼•ã„ã¦ã¿ã¾ã—ã‚‡ã†ã€‚

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(dataframe.loc[:, [2]].values,dataframe.loc[:, 3].values)

plt.figure(figsize=(5,5))
plt.scatter(dataframe.loc[:, [2]].values,dataframe.loc[:, 3].values)
plt.plot(dataframe.loc[:, 2],model.predict(dataframe.loc[:, [2]].values),color="red")
```
