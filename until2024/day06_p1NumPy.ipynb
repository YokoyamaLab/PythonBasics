### ğŸ’®[6æœ¬ç›®] å›ºæœ‰å€¤ãƒ»å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«

ç·šå½¢ä»£æ•°ã§ãŠé¦´æŸ“ã¿ã®å›ºæœ‰å€¤ãƒ»å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«ã‚‚**numpy**ãªã‚‰ç°¡å˜ã«æ±‚ã‚ã‚‹äº‹ãŒã§ãã¾ã™ã€‚ãŸã ã—ã€**numpy**ã®ã‚µãƒ–ãƒ¢ã‚¸ãƒ¥ãƒ¼ãƒ«ã§ã‚ã‚‹**linalg**ã‚’**import**ã—ãªã‘ã‚Œã°ãªã‚Šã¾ã›ã‚“ã€‚**linalg**ã¯æ…£ç¿’ã§**LA**ã«çŸ­ç¸®ãƒªãƒãƒ¼ãƒ ã—ã¦ä½¿ç”¨ã—ã¾ã™ã€‚(ãŸã ã‚Šã—ãƒªãƒ•ã‚¡ãƒ¬ãƒ³ã‚¹ç­‰ã§ã¯**linalg**ã¨æ›¸ã‹ã‚Œã¦ã„ã¾ã™ãŒã€åŒã˜ã§ã™)

````python
import numpy.linalg as LA
````

å›ºæœ‰å€¤ãƒ»å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«ã¯[linalg.eigé–¢æ•°](https://numpy.org/doc/stable/reference/generated/numpy.linalg.eig.html)ã‚’ä½¿ã„ã¾ã™ã€‚

```

```
**w**(å›ºæœ‰å€¤)ã¨**v**(å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«)ã‚’ä¸€åº¦ã«è¿”ã™ç‚¹ã«æ³¨æ„ã—ã¦ãã ã•ã„ã€‚

ãã‚Œã§ã¯ä»¥ä¸‹ã®è¡Œåˆ—ã®å›ºæœ‰å€¤ã¨å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«ã‚’æ±‚ã‚ã¾ã—ã‚‡ã†ã€‚

$$
\begin{pmatrix}

\end{pmatrix}
$$

````python
# [6] # ä¸Šè¨˜è¡Œåˆ—ã®å›ºæœ‰å€¤ãƒ»å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«ã‚’æ±‚ã‚ã‚‹
````

#### ğŸ’¯è§£ç­”ä¾‹[6æœ¬ç›®]

````python
k6 = np.array([[5,3],[4,9]])
w,v =LA.eig(k6)

print(w)
print(v)
````

##### ğŸè£œè¶³
å›ºæœ‰å€¤ãƒ™ã‚¯ãƒˆãƒ«ã¯å¤‰æ•°vã®ç¸¦æ–¹å‘ã«å…¥ã£ã¦ã„ã¾ã™ã€‚ã•ã‚‰ã«è¦æ ¼åŒ–ã‚‚ã•ã‚Œã¦ã„ã¾ã›ã‚“ã€‚

````python
print('value: ', w[0])
# å®šæ•°å€ã—ã¦ãã‚Œã£ã½ãã ã™
print('vector: ', 2 * v[:, 0] / v[1, 0])
````

### ğŸ’®[7æœ¬ç›®] ãƒ©ãƒ³ã‚¯

ã“ã‚Œã‚‚ç·šå½¢ä»£æ•°ã®åŸºæœ¬çš„ãªè¡Œåˆ—æ“ä½œã§ã™ã€‚è¡Œåˆ—ã®ãƒ©ãƒ³ã‚¯ã¯[linalg.matrix_ranké–¢æ•°](https://numpy.org/doc/stable/reference/generated/numpy.linalg.matrix_rank.html)ã‚’ä½¿ã„ã¾ã™ã€‚


### ğŸ’®[8æœ¬ç›®] é€†è¡Œåˆ—

æ‰‹è¨ˆç®—ã§ã¯3Ã—3ã§ã‚‚è‹¦åŠ´ã™ã‚‹é€†è¡Œåˆ—ã‚‚numpyã§æ±‚ã‚ã‚‰ã‚Œã¾ã™ã€‚é€†è¡Œåˆ—ã¯[linalg.invé–¢æ•°](https://numpy.org/doc/stable/reference/generated/numpy.linalg.inv.html)ã‚’ä½¿ã„ã¾ã™ã€‚

ã‚¹ãƒ©ã‚¤ã‚·ãƒ³ã‚°ã¨ã¯ãƒ‡ãƒ¼ã‚¿ã®ä¸€éƒ¨ã‚’å–ã‚Šå‡ºã™æ“ä½œã®äº‹ã§ã™ã€‚[ãƒªã‚¹ãƒˆã®æ‰€](https://github.com/YokoyamaLab/PythonBasics/blob/main/day02_01DataStructure.ipynb)ã§ã‚‚å­¦ã³ã¾ã—ãŸãŒã€ã‚‚ã†ã™ã“ã—é«˜åº¦ãªæ©Ÿèƒ½ã‚’è¦‹ã¦ã„ãã¾ã—ã‚‡ã†ã€‚
